```python
import numpy as np


train_data = np.loadtxt('./hw2_train.dat.txt')
test_data = np.loadtxt('./hw2_test.dat.txt')

train_data
```

```python
# 定義sign
def sign(value):
    return -1 if value <= 0 else 1

# decision stumps
def get_hypotheis(s, x, theta):
    h = np.zeros(x.shape)
    for idx, value in enumerate(x):
        h[idx] = s * sign(value - theta)
    return h

def get_thetas(x):
    #在頭跟尾還有每個點的中間都放個theta
    x = np.sort(x, axis=0)
    theta = np.zeros(x.shape[0] + 1)
    theta[0] = x[0] - 1
    for idx in range(len(x) - 1):
        theta[idx+1] = (x[idx] + x[idx+1])/2.
    theta[len(x)] = x[x.shape[0]-1] + 1
    return theta
```

```python
exp_time = 1
for exp_idx in range(exp_time):
    X = train_data[:,0:train_data.shape[1]-1]
    y = train_data[:,train_data.shape[1]-1 : train_data.shape[1]]
    theta_for_min_e_in = 0
    dim_for_min_e_in = 0
    s_for_min_e_in = 1
    e_in_min = 9999
    for col in range(X.shape[1] - 1):
        X_data = X[ :, col:col+1]
        thetas = get_thetas(X_data) 
        e_in = 0
        e_in_plue_s = []
        e_in_minus_s = []
        for idx, theta in enumerate(thetas):
            # s = 1
            h_plus_s = get_hypotheis(1, X_data, theta)
            h_plus_s_diff = h_plus_s - y
            # e_in = 將不同的row加起來取絕對值/(2*row的total數)
            e_in_plue_s.append((np.sum(np.absolute(h_plus_s_diff))/(len(h_plus_s_diff) * 2.)))
            
            # s = -1
            h_minus_s = get_hypotheis(-1, X_data, theta)
            h_minus_diff = h_minus_s - y
            e_in_minus_s.append((np.sum(np.absolute(h_minus_diff))/(len(h_minus_diff) * 2.)))
        min_value = min(min(e_in_plue_s), min(e_in_minus_s))
        theta = 0
        s = 1
        # 找出最小e_in的theta
        if min_value in e_in_plue_s:
            theta = thetas[e_in_plue_s.index(min_value)]
        else:
            s = -1
            theta = thetas[e_in_minus_s.index(min_value)]
        e_in += min_value
        e_in_and_theta.append((e_in, theta))
        if e_in < e_in_min:
            s_for_min_e_in = s
            e_in_min = e_in
            theta_for_min_e_in = theta
            dim_for_min_e_in = col
    #19題答案
    print("e_in = " + str(e_in_min))
    
    X_test = test_data[:,dim_for_min_e_in:dim_for_min_e_in+1]
    y_test = test_data[:,test_data.shape[1]-1 : test_data.shape[1]]
    h_min_test = get_hypotheis(s_for_min_e_in, X_test, theta_for_min_e_in)
    h_min_test_diff = h_min_test - y_test
    e_out = (np.sum(np.absolute(h_min_test_diff)))/(len(h_min_test_diff) * 2.)
    
    #20題答案
    print("e_out = " + str(e_out))
    
    #e_in = 0.25    
    #e_out = 0.355


```
